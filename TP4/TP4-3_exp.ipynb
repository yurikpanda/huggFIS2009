```python
import numpy as np
from pylab import plot,show
```

```python
def f(u):
    return np.exp((-1)*u**2)

def simpson(a,b,n):
    h = (b-a)/n
    x = np.linspace(a,b-h,n)
    fx = f(x)
    s = (1/3)*h*(f(a)+f(b)+4*np.sum(fx[1::2])+2*np.sum(fx[2:n-1:2]))
    return s
def E(n):
    x1 = np.linspace(0,3,31)
    simpsoa = list(map(lambda x: simpson(0,x,n),x1))
    plot(x1,simpsoa)
    show()
```

```python
E(1000)
```
