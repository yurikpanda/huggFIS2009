```python
import numpy as np
from pylab import plot,show
```

```python
t,v = np.loadtxt("velocities.txt", unpack=True)
def trapezoidal(t0,t1):
    I = ( 0.5 * v[t0] ) + ( 0.5 * v[t1] ) + np.sum(v[t0+1:t1])
    return I
trapezoidal(0,100)
```

```python
x = list(map(lambda t1 : trapezoidal(0,t1),np.array(t,dtype="int")))
plot(t,v)
plot(t,x)
show()
```
