```python
#3
import numpy as np
from pylab import plot,show,loadtxt

#leitura dos dados do txt 
txt = loadtxt("sunspots.txt")
x_month = txt[:,0]
y_sunspots = txt[:,1]

#primeiros 1000 pontos
while len(x_month)!= 1000:
    x_month = np.delete(x_month,1000)
    y_sunspots = np.delete(y_sunspots,1000)

#cálculo da média móvel
r = 5
soma = 0
Y = x_month * 0
for k in range(5,len(y_sunspots)-5):
    soma = 0
    for m in range(-r,r+1):
        soma += y_sunspots[k+m]
    Y[k] = soma
Y = (1/(2*r)) * Y

plot(x_month,y_sunspots)
plot(x_month,Y)
show()
```
