```python
#1
import numpy as np
from pylab import plot,show

#criação das listas theta, x e y 
theta = np.linspace(0,np.pi,1000)
x = 2 * np.cos(theta) + np.cos ( 2 * theta)
y_pos = 2 * np.sin(theta) - np.sin ( 2 * theta)

plot(x,y_pos)

show()
```

```python
#2
import numpy as np
from pylab import plot,show

theta = np.linspace(0,10*np.pi,10000)
r = theta ** 2
x = r * np.cos(theta)
y = r * np.sin(theta)

plot(x,y)
show()
```
